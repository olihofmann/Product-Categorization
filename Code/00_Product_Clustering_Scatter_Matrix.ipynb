#### Analyse der vorhandenen Features
Der Scatter Plot veranschaulicht alle Features und deren Eignung.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from sklearn.preprocessing import Imputer, StandardScaler, MinMaxScaler, RobustScaler, Normalizer
```

```python
csv = pd.read_csv("../Data/Biral_Products_csv.csv", delimiter=",", header=0)
csv.drop(columns=["Gruppenbezeichnung", "Bezeichnung", "Beschreibung", "Anwendungsgebiet"], inplace=True)
csv = csv.fillna(csv.mean())

pd.plotting.scatter_matrix(csv, figsize=(15, 15), marker="o", alpha=.2)
plt.show()
```

```python
scalers = [StandardScaler(), MinMaxScaler(), RobustScaler(), Normalizer()]
for s in scalers:
    print(type(s))
    s.fit(csv)
    data_scaled = pd.DataFrame(s.transform(csv))
    pd.plotting.scatter_matrix(data_scaled, figsize=(15, 15), marker="o", alpha=.2)
    plt.show()
    
plt.close()
```
